```python
import shutil
import os
import pandas as pd

fyp_list = pd.read_excel('FYP Final Non Submission List (1).xlsx')
fyp_list.head()
```

```python
fyp_list=fyp_list[fyp_list['Submission Status']=="Submitted"]
fyp_list.head()
```

```python
source_folder = 'C:\\Users\\Dipes\\Desktop\\FYP'
destination_folder = 'C:\\Users\\Dipes\\Desktop\\FYP 1'
```

```python
for each in os.listdir(source_folder):
    londonmet_id = int(each.split()[1])  # Extract the Londonmet ID from the filename
    if londonmet_id in fyp_list['Student ID'].values:
        f_path=os.path.join(source_folder,each)
        shutil.copy(f_path,destination_folder)
```
